```python
import pandas as pd
```

```python
# Load dataset.csv into a pandas dataframe
df = pd.read_csv('../data/dataset.csv', index_col=0)
```

```python
# Display the first 5 rows of the dataframe
df.head()
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
import tiktoken
```

```python
enc = tiktoken.encoding_for_model("gpt-4")

assert enc.decode(enc.encode("hello world")) == "hello world"
enc.encode("hello world")
```

```python
text_list = list(df["text"])[:1000]
len(text_list)
```

```python
text_list[:3]
```

```python
text_strings = " ".join(text_list)
```

```python
text_sents = sent_tokenize(text_strings)
```

```python
[word_tokenize(s) for s in text_sents]
```

```python
text_tokens = enc.encode(text_strings)
```

```python
tokens = text_strings.encode("utf-8") # raw bytes
tokens = list(map(int, tokens)) # convert to a list of integers in range 0..255 for convenience

print(f"compression ratio: {len(tokens) / len(text_tokens):.2f}X")
```
